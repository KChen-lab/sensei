```python
import pickle as pkl
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)

mpl.rcParams['figure.dpi'] = 300
```

```python
from tcga_dicts import *
```

```python
def load(cancer_type):
    with open("msi-mss-" + cancer_type + ".pkl", 'rb') as f:
        return pkl.load(f)

ucec = pd.DataFrame(load('ucec'))
coad = pd.DataFrame(load('coad'))
stad = pd.DataFrame(load('stad'))

ucec.index = augmented_cell_types
coad.index = augmented_cell_types
stad.index = augmented_cell_types
```

```python
markersize = 15

plot_cell_types = [
 'B.cells.naive',
 'B.cells.memory',
 'Plasma.cells',
 'T.cells.CD8',
 'T.cells.CD4.naive',
 'T.cells.CD4.memory.resting',
 'T.cells.CD4.memory.activated',
 'T.cells.follicular.helper',
 'T.cells.regulatory..Tregs.',
 'T.cells.gamma.delta',
 'NK.cells.resting',
 'NK.cells.activated',
 'Monocytes',
 'Eosinophils',
 'Neutrophils',
 'T.cells.all',
 'B.cells.all',
 'Nk.cells.all',
 'Macrophages.all',
 'Dendritic.cells.all',
 'Mast.cells.all'
]

x_map = {v: i for i, v in enumerate(plot_cell_types)}

cell_numbers = [100, 384, 1000, 'baseline']

offset_map = {
 100: -0.3,
 384: -0.1,
 1000: 0.1,
 'baseline': 0.3}

colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

color_map = {
 100: colors[0],
 384: colors[1],
 1000: colors[2],
 'baseline': colors[3]}

mem = {100: None,
      384: None,
      1000: None,
      'baseline': None}

fig = plt.figure(figsize=(10, 3))

ax = fig.add_subplot(1, 1, 1)

#x = [v + offset_map['baseline'] for v in x_map.values()]
#y = ucec['baseline']
#ax.scatter(x, y, marker='$u$', color = color_map['baseline'], s = 8, alpha = 0.7)

for cell_type, v in x_map.items():
    x = [v + offset_map[cell_number] for cell_number in cell_numbers]
    y = [ucec[cell_number][cell_type] for cell_number in cell_numbers]
    ax.plot(x, y, lw=1, c='gray', alpha=0.4)
    ax.scatter(x, y, marker='$u$', color=[color_map[cell_number] for cell_number in cell_numbers], s=markersize, alpha=0.7)
    
for cell_type, v in x_map.items():
    x = [v + offset_map[cell_number] for cell_number in cell_numbers]
    y = [coad[cell_number][cell_type] for cell_number in cell_numbers]
    ax.plot(x, y, lw=1, c='gray', alpha=0.4)
    ax.scatter(x, y, marker='$c$', color=[color_map[cell_number] for cell_number in cell_numbers], s=markersize, alpha=0.7)

for cell_type, v in x_map.items():
    x = [v + offset_map[cell_number] for cell_number in cell_numbers]
    y = [stad[cell_number][cell_type] for cell_number in cell_numbers]
    ax.plot(x, y, lw=1, c='gray', alpha=0.4)
    ax.scatter(x, y, marker='$s$', color=[color_map[cell_number] for cell_number in cell_numbers], s=markersize, alpha=0.7)

# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Only show ticks on the left and bottom spines
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

current_ylim = ax.get_ylim()
for v in x_map.values():
    if v % 2 == 0:
        ax.fill_between([v - 0.5, v + 0.5], current_ylim[0], current_ylim[1], facecolor='lightgray', alpha=0.2)

ax.set_ylim([0, 120])
ax.set_xticks(list(x_map.values()))
ax.tick_params(axis='x', which = 'both', labelbottom=None)
ax.set_ylabel('Estimated sample size for 80% power')
ax.set_xlim(-0.5, len(x_map) + 4)
    
# Ticks
ax.tick_params(axis='x', which = 'both', labelbottom=True)
ax.set_xticks(list(i + 0.5 for i in list(x_map.values())))
ax.set_xticklabels(list(cell_types_cosmetic_dict_short[i] for i in list(x_map.keys())), rotation=30, ha='right')

legend_elements = [mpl.lines.Line2D([0], [0], marker='.', color=color_map[i], linestyle="None", markerfacecolor=color_map[i], label=i) 
                   for i in color_map]

leg1 = ax.legend(handles=legend_elements, loc='lower right', title = 'Number of cells')

legend_elements = [mpl.lines.Line2D([0], [0], marker='$u$', color='k', linestyle="None", label='UCEC')] + \
                  [mpl.lines.Line2D([0], [0], marker='$c$', color='k', linestyle="None", markerfacecolor='k', label='COAD')] + \
                  [mpl.lines.Line2D([0], [0], marker='$s$', color='k', linestyle="None", markerfacecolor='k', label='STAD')]

ax.legend(handles=legend_elements, loc='lower right', title = 'Cancer type')
ax.add_artist(leg1)
ax.yaxis.set_minor_locator(MultipleLocator(10))
ax.yaxis.set_major_locator(MultipleLocator(20))
ax.grid(axis='y', which='both')
```
