```python
def load(filename):
    with open(filename, 'rb') as f:
        return pkl.load(f)

unpaired_primary = load("unpaired.pkl")
unpaired_recurrence = load("unpaired_recurrence.pkl")
paired_primary = load("paired.pkl")
paired_recurrence = load("paired-recurrence.pkl")
msi = load("msi-mss.pkl")

unpaired_treatment = load('gse120575_unpaired.pkl')
paired_treatment = load('gse120575_paired.pkl')


display(unpaired_primary['BLCA'])
display(unpaired_recurrence['LGG'])
display(paired_primary['BLCA'])
display(paired_recurrence['LGG'])
display(msi['COAD'])

display(paired_treatment)
display(unpaired_treatment)
```

```python
markersize = 15

## All cancer types share one figure
primary_cancer_types = [
    'BLCA',
    'BRCA',
    #'CHOL',
    'COAD',
    'ESCA',
    #'GBM',
    'HNSC',
    'KICH',
    'KIRC',
    'KIRP',
    'LIHC',
    'LUAD',
    'LUSC',
    'PRAD',
    'READ',
    'STAD',
    'THCA',
    'UCEC'
]

n_cancer_types = len(primary_cancer_types)

## Each cell type has a figure
cell_types = [
    'B.cells.all',
    'T.cells.all',
    'Nk.cells.all',
    'Monocytes',
    'Eosinophils',
    'Neutrophils',
    'Macrophages.all',
    'Dendritic.cells.all',
    'Mast.cells.all'
]

n_cell_types = len(cell_types)

## Number is represented by error bar
upper_key = 100
center_key2 = 384
center_key = 1000
lower_key = float('inf')


## Pair vs unpaired by colors
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
color_map = {'unpaired': colors[0], 'paired': colors[1]}

fig = plt.figure(figsize=(1.5 * n_cell_types, 0.3 * n_cancer_types))

axes = []

elinewidth = 1
ecolor = 'k'

for cell_type in cell_types:
    if len(axes) == 0:
        axes.append(fig.add_subplot(1, n_cell_types, 1))
    else:
        axes.append(fig.add_subplot(1, n_cell_types, len(axes) + 1, sharey = axes[0]))
        
    ax = axes[-1]
    
    for v in range(n_cancer_types):
        if v % 2 == 0:
            ax.fill_between([0, 1000], v - 0.5, v + 0.5, facecolor='lightgray', alpha=0.2)
    
    # primary
    for i, cancer_type in enumerate(primary_cancer_types):
        y = i - 0.2
        x_l = unpaired_primary[cancer_type][cell_type][lower_key]
        x = unpaired_primary[cancer_type][cell_type][center_key]
        x_u = unpaired_primary[cancer_type][cell_type][upper_key]
        if x < 200:
            ax.errorbar(x, y, xerr=[[x - x_l], [x_u - x]], 
                        color=color_map['unpaired'], ecolor=color_map['unpaired'], elinewidth=elinewidth, fmt='|', ms=6)
            x = unpaired_primary[cancer_type][cell_type][center_key2]
            ax.errorbar(x, y, color=color_map['unpaired'], fmt='|', ms=6)
        
        y = i + 0.2
        x_l = paired_primary[cancer_type][cell_type][lower_key]
        x = paired_primary[cancer_type][cell_type][center_key]
        x_u = paired_primary[cancer_type][cell_type][upper_key]
        if x < 200:
            ax.errorbar(x, y, xerr=[[x - x_l], [x_u - x]], 
                        color=color_map['paired'], ecolor=color_map['paired'], elinewidth=elinewidth, fmt='|', ms=6)
            x = paired_primary[cancer_type][cell_type][center_key2]
            ax.errorbar(x, y, color=color_map['paired'], fmt='|', ms=6)
    
    ax.set_xlim([0, 200])
    ax.set_xticks([0, 100, 200])
    ax.set_xticks([20, 40, 60, 80, 120, 140, 160, 180], True)
    #ax.set_xscale('log')
    ax.grid(axis='x', which='both', alpha=0.3)

    # Hide the right and top spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    # Only show ticks on the left and bottom spines
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')

    ax.tick_params(axis='y', which = 'both', labelleft=None)
    ax.tick_params(axis='y', which='minor', length=0)
    ax.set_xticklabels([int(i) for i in ax.get_xticks()], rotation = 90, ha="center")
    ax.set_title(cell_types_cosmetic_dict_short[cell_type], fontdict={'fontsize': 10})     
    
        
ax = axes[0]
ax.set_yticks(list(i for i in range(n_cancer_types)), True)
ax.set_yticks(list(i + 0.5 for i in range(n_cancer_types)))
ax.set_yticklabels("", va = 'center')
ax.set_yticklabels(primary_cancer_types, va = 'center', minor = True)
ax.tick_params(axis='y', which = 'both', labelleft=True)
ax.tick_params(axis='y', which='minor', length=0)
ax.set_ylim([-0.5, i + 0.5])
ax.invert_yaxis()

# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Only show ticks on the left and bottom spines
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

fig.text(0.5, 0.02, 
         "Estimated sample size to ascertain changes between normal solid tissue and primary solid tumor with 80% power", 
         ha="center", va="center")
```
