```python
import numpy as np
import scipy as sp
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.multivariate_normal([1, 2], [[1, 0.81], [0.81, 1]], 300)
```

```python
df = pd.DataFrame(x)
df.columns = ['x_1', 'x_2']
```

```python
sns.jointplot(x=df['x_1'], y=df['x_2'], kind='reg')
```

```python
sns.distplot(df['x_1'], label='x_1')
sns.distplot(df['x_2'], label='x_2')
plt.legend()
plt.xlabel('')
```

```python
df['diff'] = df['x_2'] - df['x_1']
```

```python
sns.distplot(df['diff'])
```
